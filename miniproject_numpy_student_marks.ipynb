```python
import numpy as np

np.random.seed(100)

# 30 students, 5 subjects (marks between 35 and 100)
marks = np.random.randint(35, 101, size=(30, 5))
print("Marks Data:\n", marks)

# 1. Total marks per student
student_total = np.sum(marks, axis=1)
print("\nTotal Marks Per Student:\n", student_total)

# 2. Average marks per student
student_avg = np.mean(marks, axis=1)
print("\nAverage Marks Per Student:\n", student_avg)

# 3. Total marks per subject
subject_total = np.sum(marks, axis=0)
print("\nTotal Marks Per Subject:\n", subject_total)

# 4. Hardest subject (lowest total)
hard_subject = np.argmin(subject_total)
print("\nHardest Subject Index:", hard_subject)

# 5. Topper student
topper = np.argmax(student_total)
print("Topper Student Index:", topper)

# 6. Class average
class_avg = np.mean(student_total)
print("\nClass Average:", class_avg)

# 7. Students below class average
below_avg_students = np.where(student_total < class_avg)
print("\nStudents Below Average:\n", below_avg_students)

# 8. Rank students (lowest to highest)
ranking = np.argsort(student_total)
print("\nStudent Ranking (Low to High):\n", ranking)

# 9. Distinction students (avg >= 75)
distinction = np.where(student_avg >= 75)
print("\nDistinction Students:\n", distinction)

# 10. Normalize marks (0–1 scale)
min_val = marks.min()
max_val = marks.max()
normalized_marks = (marks - min_val) / (max_val - min_val)
print("\nNormalized Marks:\n", normalized_marks)
```
