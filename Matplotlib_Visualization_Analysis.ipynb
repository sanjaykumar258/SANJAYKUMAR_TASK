```python
# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style settings
plt.style.use("seaborn-v0_8")
sns.set_theme()

# Create Sample Real-World Dataset (Sales Data)
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [1200, 1500, 1700, 1600, 1800, 2100],
    "Marketing_Spend": [300, 400, 450, 420, 480, 500],
    "Profit": [200, 300, 350, 320, 400, 450]
}

df = pd.DataFrame(data)

# LINE PLOT – Sales Trend
plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Sales"],
         marker="o",
         color="blue",
         linewidth=2,
         label="Monthly Sales")

max_sales = df["Sales"].max()
max_month = df.loc[df["Sales"].idxmax(), "Month"]

plt.annotate("Peak Sales",
             xy=(max_month, max_sales),
             xytext=("Apr", max_sales+200),
             arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()

# BAR PLOT – Profit Comparison
plt.figure(figsize=(8,5))
bars = plt.bar(df["Month"], df["Profit"], color="green")

plt.title("Monthly Profit Comparison")
plt.xlabel("Month")
plt.ylabel("Profit")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 5,
             f'{height}',
             ha='center')

plt.show()

# SCATTER PLOT – Marketing vs Sales
plt.figure(figsize=(8,5))

plt.scatter(df["Marketing_Spend"],
            df["Sales"],
            color="red",
            s=100,
            alpha=0.7)

plt.title("Marketing Spend vs Sales")
plt.xlabel("Marketing Spend")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# PIE CHART – Sales Distribution
plt.figure(figsize=(6,6))

plt.pie(df["Sales"],
        labels=df["Month"],
        autopct="%1.1f%%",
        startangle=90)

plt.title("Sales Distribution by Month")
plt.show()

# DONUT CHART
plt.figure(figsize=(6,6))

wedges, texts, autotexts = plt.pie(
    df["Sales"],
    labels=df["Month"],
    autopct="%1.1f%%",
    startangle=90
)

centre_circle = plt.Circle((0,0),0.70,fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Sales Donut Chart")
plt.show()

# HISTOGRAM – Profit Distribution
plt.figure(figsize=(8,5))

sns.histplot(df["Profit"], bins=5, kde=True)

plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()

# HEATMAP – Correlation Matrix
plt.figure(figsize=(6,5))

correlation = df[["Sales", "Marketing_Spend", "Profit"]].corr()

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            linewidths=0.5)

plt.title("Feature Correlation Heatmap")
plt.show()
```
