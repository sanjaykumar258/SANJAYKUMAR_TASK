```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Sample Data


months = np.arange(1, 13)

sales_2024 = [120, 150, 170, 160, 180, 200, 210, 230, 240, 260, 270, 300]
sales_2025 = [130, 160, 180, 170, 190, 220, 240, 250, 260, 280, 300, 320]

customer_visits = np.random.randint(50, 200, 100)
purchase_amount = np.random.randint(100, 1000, 100)

product_names = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Monitor']
product_sales = [150, 300, 120, 200, 180]


# Create Figure Layout
plt.figure(figsize=(14,10))
# 1. Line Plot (Monthly Sales)

plt.subplot(2,2,1)

plt.plot(months, sales_2024,
         marker='o',
         linestyle='-',
         linewidth=2,
         label='Sales 2024')

plt.plot(months, sales_2025,
         marker='s',
         linestyle='--',
         linewidth=2,
         label='Sales 2025')

plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()

# Find peak value for annotation
peak_value = max(sales_2025)
peak_index = sales_2025.index(peak_value)

plt.annotate(
    f'Peak: {peak_value}',
    xy=(months[peak_index], peak_value),
    xytext=(months[peak_index]-2, peak_value-40),
    arrowprops=dict(arrowstyle='->')
)
# 2. Bar Chart (Product Sales)

plt.subplot(2,2,2)

plt.bar(product_names,
        product_sales)

plt.title("Product Sales Distribution")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.grid(axis='y')

# 3. Scatter Plot (Customer Visits vs Purchase)

plt.subplot(2,2,3)

plt.scatter(customer_visits,
            purchase_amount,
            alpha=0.7,
            marker='o')

plt.title("Customer Visits vs Purchase Amount")
plt.xlabel("Number of Visits")
plt.ylabel("Purchase Amount")
plt.grid(True)
# 4. Histogram (Purchase Distribution)

plt.subplot(2,2,4)

plt.hist(purchase_amount,
         bins=10,
         edgecolor='black')

plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.grid(True)

# Adjust Layout and Show

plt.tight_layout()
plt.show()
```
