```python
# 1. Import Libraries

import numpy as np
import pandas as pd


# 2. Load Dataset

file_path = "/content/SuperStoreOrders - SuperStoreOrders.csv"
df = pd.read_csv(file_path)

print("First 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nColumn Names")
print(df.columns)


# 3. Clean Column Names

df.columns = df.columns.str.lower().str.replace(" ", "_")


# 4. Convert Numeric Columns (FIX FOR NUMPY ERROR)

df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
df['shipping_cost'] = pd.to_numeric(df['shipping_cost'], errors='coerce')
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')


# 5. Dataset Information

print("\nDataset Info")
print(df.info())

print("\nStatistical Summary")
print(df.describe())


# 6. Check Missing Values

print("\nMissing Values")
print(df.isnull().sum())


# 7. Unique Values

print("\nUnique Categories")
print(df['category'].unique())

print("\nUnique Regions")
print(df['region'].unique())

print("\nSegment Count")
print(df['segment'].value_counts())


# 8. Fix Date Format (DD/MM/YYYY)

df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True, errors='coerce')

print("\nDuplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()


# 9. NumPy Operations

sales_array = df['sales'].to_numpy()
profit_array = df['profit'].to_numpy()
shipping_array = df['shipping_cost'].to_numpy()

print("\nNumPy Analysis")

print("Average Sales:", np.mean(sales_array))
print("Maximum Sales:", np.max(sales_array))
print("Minimum Sales:", np.min(sales_array))

print("Average Profit:", np.mean(profit_array))
print("Maximum Profit:", np.max(profit_array))
print("Minimum Profit:", np.min(profit_array))

print("Profit Standard Deviation:", np.std(profit_array))
print("Median Profit:", np.median(profit_array))

print("90th Percentile Profit:", np.percentile(profit_array, 90))


# 10. Total Sales and Profit

total_sales = df['sales'].sum()
total_profit = df['profit'].sum()

print("\nTotal Sales:", total_sales)
print("Total Profit:", total_profit)


# 11. Category Wise Sales

category_sales = df.groupby('category')['sales'].sum()

print("\nCategory Wise Sales")
print(category_sales)


# 12. Region Wise Profit

region_profit = df.groupby('region')['profit'].sum()

print("\nRegion Wise Profit")
print(region_profit)


# 13. Segment Wise Sales

segment_sales = df.groupby('segment')['sales'].sum()

print("\nSegment Wise Sales")
print(segment_sales)


# 14. Top 10 Products by Sales

top_products = df.sort_values(by='sales', ascending=False).head(10)

print("\nTop 10 Products by Sales")
print(top_products[['product_name', 'sales']])


# 15. Discount Impact on Profit

discount_profit = df.groupby('discount')['profit'].mean()

print("\nDiscount vs Average Profit")
print(discount_profit)


# 16. Shipping Cost Analysis

avg_shipping = df['shipping_cost'].mean()
max_shipping = df['shipping_cost'].max()
min_shipping = df['shipping_cost'].min()

print("\nShipping Cost Analysis")

print("Average Shipping Cost:", avg_shipping)
print("Maximum Shipping Cost:", max_shipping)
print("Minimum Shipping Cost:", min_shipping)


# 17. Filtering Data

high_profit_orders = df[df['profit'] > 500]

print("\nHigh Profit Orders (>500)")
print(high_profit_orders[['product_name', 'profit']])


# 18. Create New Column

df['profit_margin'] = df['profit'] / df['sales']

print("\nProfit Margin Column")
print(df[['sales', 'profit', 'profit_margin']].head())


# 19. Region with Highest Sales

region_sales = df.groupby('region')['sales'].sum()

highest_region = region_sales.idxmax()

print("\nRegion with Highest Sales:", highest_region)


# 20. Year Wise Sales

df['year'] = df['order_date'].dt.year

year_sales = df.groupby('year')['sales'].sum()

print("\nYear Wise Sales")
print(year_sales)


# 21. Sort Data by Profit

sorted_profit = df.sort_values(by='profit', ascending=False)

print("\nTop Profitable Orders")
print(sorted_profit[['product_name', 'profit']].head())


# 22. Save Processed Dataset

df.to_csv("/content/processed_superstore_data.csv", index=False)

print("\nProcessed dataset saved successfully")
```
