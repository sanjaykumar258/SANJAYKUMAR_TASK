```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('/content/ultimate_student_productivity_dataset_5000.csv')

# 1. REMOVE DUPLICATES
df = df.drop_duplicates()

# 2. STANDARDIZE TEXT (Casing and Trimming)
categorical_cols = ['gender', 'academic_level', 'internet_quality']
for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# 3. HANDLE MISSING VALUES (Imputation)
# Fill numeric with median, categorical with mode
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# 4. OUTLIER HANDLING (Using IQR Method)
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Applying outlier removal to a key column (e.g., study_hours)
df = remove_outliers(df, 'study_hours')

# 5. DATA TYPE CONVERSION
df['student_id'] = df['student_id'].astype(str)

# 6. LOGICAL SANITY CHECK
# Example: Ensure sleep hours are within a humanly possible range
df = df[(df['sleep_hours'] >= 2) & (df['sleep_hours'] <= 16)]

print("Data Cleaning Complete!")
print(df.info())
```
