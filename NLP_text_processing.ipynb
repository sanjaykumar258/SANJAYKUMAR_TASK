```python
import spacy

# Load English NLP model
nlp = spacy.load("en_core_web_sm")

# Sample Text
text = """
Apple Inc. is planning to open a new office in Bangalore, India in 2026.
The company CEO Tim Cook announced the expansion during a conference in California.
"""

# Process text
doc = nlp(text)

# 1. Sentence Extraction
print("Sentences:")
for sent in doc.sents:
    print("-", sent.text)

print("\n------------------")

# 2. Keyword Extraction (Nouns)
print("Keywords:")
keywords = [token.text for token in doc if token.pos_ in ["NOUN", "PROPN"]]
print(keywords)

print("\n------------------")

# 3. Named Entity Extraction
print("Named Entities:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```
